Part 1 - Geometric Distribution

$P(X=n)=(1-p)^{n-1}p, \quad n=1,2,...$

$X = \lfloor\frac{\log(U)}{\log{(1-p)}}\rfloor +1, \quad U \sim U(0,1)$

```python
p_small = 0.1
p_medium = 0.5
p_large = 0.9
```

```python
import random
import math

def geometric_sample(p):
    u = random.random()
    return int(math.log(u) / math.log(1-p)) + 1

N = 10000

sample_small = []
sample_medium = []
sample_large = []

for _ in range(N):
    sample_small.append(geometric_sample(0.1))
    sample_medium.append(geometric_sample(0.5))
    sample_large.append(geometric_sample(0.9))
```

```python
def geometric_pmf(n, p):
    return ((1-p)**(n-1))*p
```

```python
def freq(data):

    counts = {}

    for x in data:
        if x not in counts:
            counts[x] = 0
        counts[x] +=1
    
    return counts
```

```python
frequency_small = freq(sample_small)
frequency_medium = freq(sample_medium)
frequency_large = freq(sample_large)

for x in sorted(frequency_small):
    frequency_small[x] /= len(sample_small)

for x in sorted(frequency_medium):
    frequency_medium[x] /= len(sample_medium)

for x in sorted(frequency_large):
    frequency_large[x] /= len(sample_large)
```

```python
print("x : Emperical vs. Theoretical, p = 0.1")
for i in range(1, 7):
    print(i, ": {:.4f}".format(frequency_small[i]), "\t", "{:.4f}".format(geometric_pmf(i, 0.1)))
```

```python
print("x : Emperical vs. Theoretical, p = 0.5")
for i in range(1, 7):
    print(i, ": {:.4f}".format(frequency_medium[i]), "\t", "{:.4f}".format(geometric_pmf(i, 0.5)))
```

```python
print("x : Emperical vs. Theoretical, p = 0.9")
for i in range(1, 5):
    print(i, ": {:.4f}".format(frequency_large[i]), "\t", "{:.4f}".format(geometric_pmf(i, 0.9)))
```

```python
# change n to check if it fits with probability
n = 1

expected_small = N * geometric_pmf(n, p_small)
expected_medium = N * geometric_pmf(n, p_medium)
expected_large = N * geometric_pmf(n, p_large)
expected_small, expected_medium, expected_large
```

Small p = 0.1 -> long right tail, large values occur frequently
Medium p = 0.5 -> ~75% of observations are 1 or 2
Large p = 0.9 -> most observations are 1

Part 2 - Six-point distribution

Crude method

$F(x_{i-1}) \lt U\leq F(x_i)$

```python
values = [1,2,3,4,5,6]

probs = [
    7/48,
    5/48,
    1/8,
    1/16,
    1/4,
    5/16
]

# total = 0
# for i in probs:
#     total += i
# total
```

```python
cdf = []

running = 0

for p in probs:
    running += p
    cdf.append(running)
```

```python
def direct_sample():

    u = random.random()

    for i in range(len(cdf)):
        if u <= cdf[i]:
            return values[i]
```

```python
N = 10000

res = {}

for i in range(N):
    temp = direct_sample()

    if temp not in res:
        res[temp] = 0
    res[temp] += 1
```
