```python
import numpy as np
```

```python
def bootstrap_replicates(x, estimator, k):
    replicates = []
    for _ in range(k):
        sample = np.random.choice(x, size=len(x))
        replicates.append(estimator(sample))
    return np.array(replicates)
```

```python
# part 1

x = np.array([56, 101, 78, 67, 93, 87, 64, 72, 80, 69])
a = -5
b = 5
k = 1000

x_mean = x.mean()
reps = bootstrap_replicates(x, np.mean, k) - x_mean

count = 0
for replicate in reps:
    if a < replicate < b:
        count += 1

print(x_mean)
print(count/k)
```

```python
# part 2

x = np.array([5, 4, 9, 6, 21, 17, 11, 20, 7, 10, 21, 15, 13, 16, 8])

reps = bootstrap_replicates(x, lambda x: x.var(ddof=1))
var_s2 = reps.var(ddof=1)

print(var_s2)
```

```python
# part 3

beta, k_pareto, n = 1, 1.05, 200
u = np.random.uniform(size=n)
pareto_sample = beta / u ** (1 / k_pareto)

print(pareto_sample.mean())
print(np.median(pareto_sample))

var_mean = bootstrap_replicates(pareto_sample, np.mean).var(ddof=1)
var_median = bootstrap_replicates(pareto_sample, np.median).var(ddof=1)

print(var_mean)
print(var_median)
```
