Part 1 - Ross Exercise 13

```python
import numpy as np

x = np.array([
    56,101,78,67,93,
    87,64,72,80,69
])

n = len(x)

xbar = np.mean(x)

a = -5
b = 5

B = 10000

count = 0

for _ in range(B):

    sample = np.random.choice(
        x,
        size=n,
        replace=True
    )

    mean_star = np.mean(sample)

    if a < mean_star - xbar < b:
        count += 1

p_hat = count / B

print("Bootstrap estimate:", p_hat)
```

Part 2 - Ross Exercise 15

```python
import numpy as np

x = np.array([
    5,4,9,6,21,
    17,11,20,7,10,
    21,15,13,16,8
])

n = len(x)

B = 10000

bootstrap_variances = []

for _ in range(B):

    sample = np.random.choice(
        x,
        size=n,
        replace=True
    )

    s2 = np.var(
        sample,
        ddof=1
    )

    bootstrap_variances.append(s2)

var_s2 = np.var(
    bootstrap_variances,
    ddof=1
)

print(
    "Bootstrap estimate of Var(S²):",
    var_s2
)
```

Part 3.a - Pareto Sample

```python
import numpy as np

n = 200

beta = 1
k = 1.05

u = np.random.rand(n)

x = beta * u**(-1/k)

sample_mean = np.mean(x)
sample_median = np.median(x)

print("Mean:", sample_mean)
print("Median:", sample_median)
```

Part 3.b - Bootstrap variance of sample mean

```python
B = 100

bootstrap_means = []

for _ in range(B):

    sample = np.random.choice(
        x,
        size=n,
        replace=True
    )

    bootstrap_means.append(
        np.mean(sample)
    )

mean_variance = np.var(
    bootstrap_means,
    ddof=1
)

print(
    "Bootstrap variance of mean:",
    mean_variance
)
```

Part 3.c - Bootstrap variance of sample median

```python
bootstrap_medians = []

for _ in range(B):

    sample = np.random.choice(
        x,
        size=n,
        replace=True
    )

    bootstrap_medians.append(
        np.median(sample)
    )

median_variance = np.var(
    bootstrap_medians,
    ddof=1
)

print(
    "Bootstrap variance of median:",
    median_variance
)
```

Part 3.d - Precision comparison

```python
# Pareto distribution with k=1.05 
print(
    "Variance of mean:",
    mean_variance
)

print(
    "Variance of median:",
    median_variance
)
```
