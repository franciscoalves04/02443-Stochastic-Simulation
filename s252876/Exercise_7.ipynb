```python
def TSP(cost_matrix):
    # A is cost matrix (n,n), with A[i,j] the cost of traveling from city i to city j
    
    def T(k):
        return 1 / np.sqrt(1+k)
    
    def C(r):
        i = r
        j = np.concatenate([r[1:] , r[0:1]])
        route_index = np.column_stack([i,j])
        c = np.sum(A[route_index[:, 0], route_index[:, 1]])
        
        
        return c
    
    def proposal(r):
        i, j = np.random.choice(n, size=2, replace=False)
        proposal_route = r.copy()
        proposal_route[i], proposal_route[j] = proposal_route[j], proposal_route[i]
        
        return proposal_route
        
        
    
    A = cost_matrix.copy()
    k = 0
    n = A.shape[0] # n cities
    np.fill_diagonal(A, A.max()+1)
    iterations = 100
    
    route = np.arange(0, n)
    print("Inital route=", route)
    print("Initial cost=",C(route))
    
    for k in range(iterations):
        proposal_r = proposal(route)
        delta_C = C(proposal_r) - C(route)
        alpha = min(1, np.exp(-delta_C / T(k)))
        if np.random.rand() < alpha:
            route = proposal_r
    
    
    print("End route=", route)
    print("End cost=",C(route))
    
    
    return route, C(route)
    
    
    
    
```
